````python
from pathlib import Path\n
import hashlib\n
import json\n
\n
wish_id = 'wish.replace_me'\n
out_dir = Path('artifacts/wishes') / wish_id\n
out_dir.mkdir(parents=True, exist_ok=True)\n
\n
mmd = '\n'.join([\n
    'flowchart TD',\n
    '  START[STATE: START] --> READY[STATE: READY]',\n
    '  START --> NEED_INFO[STATE: NEED_INFO]',\n
    '  READY --> EXECUTED[STATE: EXECUTED]',\n
    '  EXECUTED --> VERIFIED[STATE: VERIFIED]',\n
    '  EXECUTED --> FAIL_CLOSED[STATE: FAIL_CLOSED]',\n
]) + '\n'\n
\n
state_path = out_dir / 'state.mmd'\n
sha_path = out_dir / 'state.sha256'\n
results_path = out_dir / 'results.json'\n
\n
state_path.write_text(mmd, encoding='utf-8')\n
digest = hashlib.sha256(mmd.encode('utf-8')).hexdigest()\n
sha_path.write_text(f'{digest}  state.mmd\n', encoding='utf-8')\n
\n
results = {\n
    'wish_id': wish_id,\n
    'status': 'TEMPLATE',\n
    'artifacts': {\n
        'state_mmd': str(state_path),\n
        'state_sha256': str(sha_path),\n
    },\n
}\n
results_path.write_text(json.dumps(results, indent=2) + '\n', encoding='utf-8')\n
print(f'Wrote: {state_path}')\n
print(f'Wrote: {sha_path}')\n
print(f'Wrote: {results_path}')
````
