# HOW-TO-SECRET-SAUCE-INTEGRATION

Quest: Game of Death Integration
Belt target: Green -> Brown

Goal:
- map root secret-sauce notebooks to executable CLI commands
- verify parity via tests and receipts

Gate:
- parity test file exists
- required commands execute successfully
- recipe/skills/books views expose required artifacts


```python
from pathlib import Path
import json
import subprocess

repo = Path.cwd().resolve()
while repo != repo.parent and not (repo / "cli" / "src").exists():
    repo = repo.parent
assert (repo / "cli" / "src").exists(), "repo root not found"

def run(cmd: str) -> None:
    print("\n$ " + cmd)
    subprocess.run(cmd, shell=True, check=True, cwd=repo)

run("PYTHONPATH=cli/src python -m stillwater recipe list --dir cli/recipes --json")
run("PYTHONPATH=cli/src python -m stillwater books list --dir books --json")
run("PYTHONPATH=cli/src python -m stillwater skills-ab --backend mock --run-id howto-secret-sauce")
run("PYTEST_DISABLE_PLUGIN_AUTOLOAD=1 pytest -q cli/tests/test_notebook_root_parity.py")

print("Secret-sauce integration checks passed.")
```
