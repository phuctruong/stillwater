# HOW-TO-WISH-NOTEBOOK-LOOP

Quest: Wish Engine
Belt target: Yellow -> Brown

Rule:
- wish is the contract
- notebook is the execution proof
- receipts are the promotion gate

Gate:
- wish artifacts validate
- notebook run is clean and repeatable


```python
from pathlib import Path
import subprocess

repo = Path.cwd().resolve()
while repo != repo.parent and not (repo / 'cli' / 'src').exists():
    repo = repo.parent
assert (repo / 'cli' / 'src').exists(), 'Repo root not found'

wish_files = sorted((repo / 'cli' / 'wishes').glob('wish.*.md'))
notebooks = sorted((repo / 'cli' / 'notebooks').glob('HOW-TO-*.ipynb'))
print('wishes:', len(wish_files))
print('notebooks:', len(notebooks))
assert wish_files, 'No wish documents found in cli/wishes'
assert notebooks, 'No HOW-TO notebooks found in cli/notebooks'

cmd = 'PYTHONPATH=' + str(repo / 'cli/src') + ' python -m stillwater wish list --json'
print('\n$ ' + cmd)
p = subprocess.run(['bash', '-lc', cmd], cwd=str(repo), capture_output=True, text=True)
print(p.stdout)
if p.returncode != 0:
    print(p.stderr)
    raise RuntimeError('Command failed (' + str(p.returncode) + '): ' + cmd)
```
