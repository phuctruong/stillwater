```python
from pathlib import Path
import json
import subprocess

repo = Path.cwd().resolve()
while repo != repo.parent and not (repo / 'cli' / 'src').exists():
    repo = repo.parent
assert (repo / 'cli' / 'src').exists(), 'Repo root not found'

cmd = (
    'PYTHONPATH=' + str(repo / 'src/cli/src')
    + ' python -m stillwater skills-ab --backend mock --no-cache --run-id convention_density_demo'
)
print('$ ' + cmd)
p = subprocess.run(['bash', '-lc', cmd], cwd=str(repo), capture_output=True, text=True)
print(p.stdout)
if p.returncode != 0:
    print(p.stderr)
    raise RuntimeError('Command failed (' + str(p.returncode) + '): ' + cmd)

results = repo / 'artifacts' / 'skills_ab' / 'results.json'
assert results.exists(), 'results.json not found after skills-ab run'
data = json.loads(results.read_text(encoding='utf-8'))
print('backend:', data.get('backend'))
print('model:', data.get('model'))
print('scenarios:', len(data.get('results', [])))
```
