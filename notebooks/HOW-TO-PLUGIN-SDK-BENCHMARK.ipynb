```python
from pathlib import Path

repo = Path.cwd().resolve()
while repo != repo.parent and not (repo / 'cli' / 'src').exists():
    repo = repo.parent
assert (repo / 'cli' / 'src').exists(), 'Repo root not found'

paths = [repo / 'cli' / 'src' / 'stillwater', repo / 'skills', repo / 'cli' / 'tests']
for p in paths:
    exists = p.exists()
    count = len(list(p.rglob('*'))) if exists else 0
    print(p, 'files=', count)
    if not exists or count == 0:
        raise RuntimeError('Path missing or empty: ' + str(p))
```
